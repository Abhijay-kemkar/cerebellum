```python
import numpy as np
import json
import matplotlib.pyplot as plt

folder = '/home/srujanm/snemi/zhuhd15_scripts/'
base = 'aff85_his256'
Tthres2 = np.array([0.10,0.15,0.20,0.25,0.30,0.35,0.40])
n_trials = Tthres2.size
rand_merge = np.zeros(n_trials)
rand_split = np.zeros(n_trials)
voi_merge = np.zeros(n_trials)
voi_split= np.zeros(n_trials)

# Using results previously generated by waterz
for i in range(n_trials):
    filepath = folder+base+'_%.2f.json' %(Tthres2[i])
    with open(filepath) as f:
        data = json.load(f)
    rand_merge[i] = 1.0-float(data['rand_merge'])
    rand_split[i] = 1.0-float(data['rand_split'])
    voi_merge[i] = float(data['voi_merge'])
    voi_split[i] = float(data['voi_split'])
```

```python
fig1 = plt.figure()
plt.scatter(rand_merge, rand_split)
plt.title('zwatershed thresh sweep')
plt.xlabel('Merge error rand')
plt.ylabel('Split error rand')
plt.show()

fig2 = plt.figure()
plt.scatter(voi_merge, voi_split)
plt.title('zwatershed thresh sweep')
plt.xlabel('VOI merge')
plt.ylabel('VOI split')
plt.show()
```
