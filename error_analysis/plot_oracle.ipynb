```python
# plots delta_VI results from oracle-ifying individual segments in GT with poor IoU 
# see IoU_rank.py for exact implementation
import numpy as np

#pre-sorted in decreasing order of VI sort + VI merge
delta_VI = np.load('/home/srujanm/snemi/analysis_results/delta_total.npy')
```

```python
# plot stacked bar chart with merge and split VI changes
import matplotlib as mpl
import matplotlib.pyplot as plt
% matplotlib inline

n_segs = delta_VI.shape[0]
ind = np.arange(n_segs)
width = 0.5

fig, ax = plt.subplots(1)
psplit = ax.bar(ind, delta_VI[:,1], width, color='#d62728')
pmerge = ax.bar(ind, delta_VI[:,2], width, bottom=delta_VI[:,1])

plt.ylabel('delta_VI')
plt.xlabel('GT id of segment')
plt.title('Objects with highest delta_VI')
plt.xticks(ind, delta_VI[:,0].astype(int))
plt.legend((psplit[0], pmerge[0]), ('delta_VI split', 'delta_VI merge'))
fig.set_size_inches(16, 6)

plt.show()
```
